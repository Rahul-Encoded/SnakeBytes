```python
import random

questions = (
	"What is the output of print(2 ** 3)?",
	"How do you create a list in Python?",
	"What is a lambda function in Python?",
	"What is the difference between a list and a tuple in Python?",
	"How do you handle exceptions in Python?",
)

options = (
	("A. 8", "B. 6", "C. 9", "D. 7"),
	("A. Using []", "B. Using {}", "C. Using ()", "D. Using <>"),
	("A. A function with a name", "B. A function with no name", "C. A function with multiple names", "D. None of the above"),
	("A. Both are immutable", "B. Lists are immutable, tuples are mutable", "C. Both are mutable", "D. Lists are mutable, tuples are immutable"),
	("A. Using loops", "B. Using if and else blocks", "C. Using try and except blocks", "D. None of the above"),
)

answers = ("A", "A", "B", "D", "C")
question_num = 0
guesses = []
score = 0


for que in questions:
	print(que)
	for option in options[question_num]:
		print(option)
	guess = input(f"Enter your option key:").upper()
	guesses.append(guess)
	if(guesses[question_num] == answers[question_num]):
		score += 1
		print(f"Correct!\n")
	else:
		print(f"Wrong!\n")
	question_num += 1

print(f"\nYour total correct guesses are: {score}\n")
print("Following were your guesses and their respective answers: \n")
print("Your guess <=> Actual answer")
for num in range(len(answers)):
	print(f"{guesses[num]:^10} <=> {answers[num]:^10}")
```
