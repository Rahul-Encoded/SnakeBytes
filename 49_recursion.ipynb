recursion = a function that calls itself from within
helps to visualize a complex problem into basic steps,
problems can be solved more easily iteratively or recursively

iterative = faster, complex

recursive = slower, simpler

```python
import time

# ITERATIVE
def walk_iterative(steps):
    for step in range(1, steps + 1):
        print(f"You take step #{step}")

# RECURSIVE
def walk_recursive(steps):
    if steps == 0:
        return
    walk_recursive(steps - 1)
    print(f"You take step #{steps}")
```

```python
# Measure time for the iterative function
start_time = time.time()
walk_iterative(100)
end_time = time.time()
print(f"Iterative walk took {end_time - start_time:.6f} seconds")
```

```python
# Measure time for the recursive function
start_time = time.time()
walk_recursive(100)
end_time = time.time()
print(f"Recursive walk took {end_time - start_time:.6f} seconds")
```

```python
# ITERATIVE
def factorial(x):
    result = 1
    if x > 0:
        for i in range(1, x + 1):
            result *= i
        return result

print(factorial(10))
# RECURSIVE
def factorial(x):
    if x == 1:
        return 1
    else:
        return x * factorial(x - 1)

print(factorial(10))
```
