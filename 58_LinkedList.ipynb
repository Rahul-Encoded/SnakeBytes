```python
# Operations that can be performed on a linked list:
# 1. Insertion: Add a new node to the linked list.
#    - At the beginning
#    - At the end
#    - After a given node
# 2. Deletion: Remove a node from the linked list.
#    - From the beginning
#    - From the end
#    - A specific node by value
# 3. Traversal: Visit each node in the linked list.
# 4. Search: Find a node in the linked list by value.
# 5. Update: Modify the value of a node in the linked list.
# 6. Reverse: Reverse the order of nodes in the linked list.
# 7. Length: Calculate the number of nodes in the linked list.
# 8. Detect Loop: Check if there is a cycle in the linked list.
# 9. Merge: Combine two linked lists into one.
# 10. Sort: Arrange the nodes in the linked list in a specific order.
```

```python
class Node:
	def __init__(self, data=None, next=None):
		self.data = data
		self.next = next


class LL:
	def __init__(self):
		self.head = None

	#insert at the beginning
	def insert_at_start(self, data):
		node = Node(data, self.head)
		self.head = node
		# if(self.head == None):
		# 	self.head = node
		# else:
		# 	node.next = self.head
		# 	self.head = node
	
	#insert at the end
	def insert_at_end(self, data):
		node = Node(data)
		if self.head is None: 
			self.head = node
		else:
			ptr = self.head
			while ptr.next:
				ptr = ptr.next
			ptr.next = node
	
	#insert after a given node
	def insert(self, data, index):
		node = Node(data)
		ptr = self.head
		count = 0
		while count < index:
			if (ptr.next is None):
				print("Enter a valid index!")
				break
			pre = ptr
			ptr = ptr.next
			count += 1
		node.next = ptr
		pre.next = node

	
	def search_by_value(self, data):
		ptr = self.head
		count = 0
		while ptr:
			if(ptr.data == data):
				return (f"Your element {data} is at index {count}")
			else:
				return (f"No matching elements with {data}")
			count += 1
	
	def search_by_index(self, index):
		ptr = self.head
		count = 0
		while count < index:
			if ptr.next is None:
				return (f"Enter a valid index. You linked list size is {count}")
	
			elif(count == index):
				return (f"Your data at index {index} is {ptr.data}")
			ptr = ptr.next
			count += 1
		

	def printLL(self):
		if (self.head == None):
			print("Linked List is empty")
		else:
			ptr = self.head
			while (ptr):
				print(f"{ptr.data} -->", end=" ")
				ptr = ptr.next

```
