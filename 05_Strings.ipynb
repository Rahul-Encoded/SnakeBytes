```python
#useful string methods
name = input("Enter your full name: ")
ph_num = input("Enter your phone number: ")

result = len(name)
print(result)

result = name.find(" ") #for first occurence
print(result)

result = name.rfind(" ") #for last occurence, r means reverse
print(result)

result = name.capitalize()
print(result)

result = name.upper()
print(result)

result = name.lower()
print(result)

result = name.isdigit() #all the char are digits
print(result)

result = name.isalpha() #for alphabetical char only
print(result)

result = ph_num.count("-")
print(result)

result = ph_num.replace("-", " ")
print(result)

#for the list of all the string methods just type
print(help(str))
```

```python
# #validate user input exercise 
# 1. username is no more than 12 characters
# 2. username must not contain spaces
# 3. username must not contain digits
```

```python
username = input("Enter a valid username: ")

def error():
    print("Enter a valid username!!!")

if(len(username) > 12):
    error()
if(username.isalpha() == False):
    error()
else:
    print("Username accepted")
```

```python
#string indexing: 
# accessing elements of a sequence using [] (indexing operator)
# [start:end:step] 
# start is inclusive
# end is exclusive

credit_number = "1234-5678-9012-3456"

result = credit_number[0]
print(result)

result = credit_number[0:4] #[:4]
print(result)

print(credit_number[4:8])
print(credit_number[4:])
print(credit_number[-1])
print(credit_number[-4:])#last 4 digit exercise
print(credit_number[::2])
print(credit_number[::3])
```

```python
#Reverse the string
result = credit_number[::-1]
print(result)
```

```python
result = reversed(credit_number)
print(result)
```

```python
result = ''.join(reversed(result))
print(result)
```

```python
# Email Slicer Program

email = input("Enter your email")

index = email.find("@")
email_client=email[:index]
email_server=email[index+1:]

print(f"Email client = {email_client} \nEmail Server = {email_server}")
```
